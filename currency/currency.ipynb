```python
import requests
import pandas as pd
import sched
import time
import json
import datetime
```

```python
r=requests.get('https://tw.rter.info/capi.php')
currency=r.json()
print('時間:',currency['USDTWD']['UTC'],'匯率:',currency['USDTWD']['Exrate'])
```

```python
s = sched.scheduler(time.time, time.sleep)
def test():
    time = datetime.datetime.now()
    r=requests.get('https://tw.rter.info/capi.php')
    currency=r.json()
    print('實際時間:',time.strftime("%Y-%m-%d %H:%M:%S"),'匯率時間:',currency['USDTWD']['UTC'],'匯率:',currency['USDTWD']['Exrate'])
    start_time = datetime.datetime.strptime(str(time.date())+'14:30', '%Y-%m-%d%H:%M')
    end_time =  datetime.datetime.strptime(str(time.date())+'17:50', '%Y-%m-%d%H:%M')
    
    # 判斷爬蟲終止條件
    if time >= start_time and time <= end_time:
        s.enter(60, 0, test,argument=() )#(等待秒數,優先順序,函數,參數)
    #return c
```

```python
s.enter(0, 0, test,argument=())
s.run()
```
