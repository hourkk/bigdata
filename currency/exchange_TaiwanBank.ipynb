```python
#
#抓取台灣銀行牌告匯率 by 大數學堂
#http://www.largitdata.com

import pandas as pd
from datetime import datetime
import sqlite3

#source
url="http://rate.bot.com.tw/xrt?Lang=zh-TW"

#讀取網頁
dfs=pd.read_html(url)

#測試
#type(dfs)
#len(dfs)

#資料呈現
#dfs[0]

currency=dfs[0]

```

```python
#檢查
#type(currency)

#擷取需要的欄位
currency_fix=currency.ix[:,0:5]

#檢查
#currency_fix

#自訂欄位名稱
currency_fix.columns=[u'幣別',u'現金匯率-本行買入',u'現金匯率-本行賣出',u'即期匯率-本行買入',u'即期匯率-本行賣出']

#檢查
#currency_fix

#清除幣別欄重複字元
currency_fix[u'幣別']=currency_fix[u'幣別'].str.extract('\((\w+)\)')

#檢查
currency_fix

#存檔
#currency_fix.to_excel('currency.xlsx')
#currency_fix.to_csv('currency.csv')
```

```python
#將資料存到Sqlite

#新增日期欄位&使用今天的日期
currency_fix['Date']=datetime.now().strftime('%Y-%m-%d')
#檢查格式
currency_fix.info()
```

```python
#將Data欄位用pandas轉成日期的資料型態
currency_fix['Date']=pd.to_datetime(currency_fix['Date'])
```

```python
#檢查格式
currency_fix.info()
```

```python
#印出&檢查內容
currency_fix
```

```python
#寫檔
with sqlite3.connect('currency.sqlite') as db:
    currency_fix.to_sql('currency',con=db,if_exists='append')
```

```python
#讀檔(在jupyter上直接看)
with sqlite3.connect('currency.sqlite') as db:
    df=pd.read_sql_query('select * from currency', con=db)
df
```
